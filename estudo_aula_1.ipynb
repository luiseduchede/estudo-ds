```python
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
dados.sample(15)
```

```python
#1 - Realizar a média de cada um dos bairros
grupo = dados.groupby(["Bairro"])["Metragem"].mean()
grupo
```

```python
#2 - Duas formas de selecionar dados por bairro
#2.1
bairro = "Aclimação"
ex = dados[dados["Bairro"] == bairro]
ex
```

```python
#2.2
ex = dados.iloc[:, 1]
ex
```

```python
#3 - Explorar alguns gráficos do Pandas
ex = dados["Bairro"].value_counts()
ex.head(10).plot.pie(subplots=True, autopct="%i", figsize=(8, 4))
```

```python
ex = dados.groupby(["Bairro"])["Metragem"].mean().head(100).plot.hist()
```

```python
ex = dados.iloc[:, [3, 4]]
ex.head(50).plot.hexbin(x = "Metragem", y = "Quartos", gridsize = 20)
```
